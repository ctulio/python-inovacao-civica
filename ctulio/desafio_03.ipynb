### Acesse a api da camara, e recupere os reembolsos do ano de 2018

```python
import requests

# fazendo o download do conteúdo do arquivo
url = f"http://www.camara.leg.br/cotas/Ano-2018.csv.zip"
r = requests.get(url)
```

### Limpe a coluna `txtCNPJCPF`: ela deve conter somente dígitos numéricos

```python
df_reimbursements.txtCNPJCPF.head()
```

```python
# regex 
# \d casa com caracteres que sejam dígitos
# \D casa com caracteres que não sejam dígitos (equivalente a [^\d], lista que nega caracteres dígitos)
df_reimbursements.txtCNPJCPF = df_reimbursements.txtCNPJCPF.replace(regex=[r'[^\d]'], value='')

# código da aula
#df_reimbursements.txtCNPJCPF = df_reimbursements.txtCNPJCPF.str.replace(r'[^\d]', '', regex=True)
```
