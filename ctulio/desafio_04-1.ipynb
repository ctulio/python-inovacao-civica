### Consulte se houveram gastos reembolsados associados a empresas que na época estavam com status `08`


situation == 8 significa `BAIXADA` (*a empresa foi extinta, seja por solicitação dos sócios ou por ação da Receita*)

```python
diaSemana = {0:'segunda',
            1:'terça',
            2:'quarta',
            3:'quinta',
            4:'sexta',
            5:'sábado',
            6:'domingo'}

def classificaDia(row):
    if row == 0:
        return 'segunda'
    elif row == 1:
        return 'terça'
    elif row == 2:
        return 'quarta'
    elif row == 3:
        return 'quinta'
    elif row == 4:
        return 'sexta'
    elif row == 5:
        return 'sábado'
    else:
        return 'domingo'
```

```python
df['codDiaSemana'] = df['datEmissao'].dt.dayofweek#.apply(classificaDia)
```

```python
df['diaSemana'] = df['datEmissao'].dt.dayofweek.apply(classificaDia)
```
